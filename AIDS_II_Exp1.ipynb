```python
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/AIDS Dataset/dengue.csv')

data
```

```python
city = data['dengue.residence']
city.head()
```

```python
no_of_days_dengue = data['dengue.days'].astype('str')
no_of_days_dengue.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,6))
plt.plot(city, no_of_days_dengue);

plt.xlabel('city')
plt.ylabel('days')
```

```python
plt.hist(data['dengue.current_temp'])
```

```python
plt.hist(data['dengue.dengue'].astype('str'))
```

```python
import plotly.express as px

fig = px.scatter_matrix(data, dimensions=['dengue.residence', 'dengue.days', 'dengue.dengue'])
fig.show()
```
