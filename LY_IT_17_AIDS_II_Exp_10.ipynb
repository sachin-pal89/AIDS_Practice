```python
import networkx as nx
import matplotlib.pyplot as plt

def dijkstra(graph, start):
    # Create a dictionary to store the shortest distances from the start node to other nodes
    shortest_distances = {node: float('inf') for node in graph}
    shortest_distances[start] = 0

    # Create a dictionary to store the previous node in the shortest path
    previous_nodes = {}

    # Create a set to keep track of visited nodes
    unvisited_nodes = set(graph)

    while unvisited_nodes:
        # Select the node with the smallest distance from the start node
        current_node = min(unvisited_nodes, key=lambda node: shortest_distances[node])

        # Remove the current node from the unvisited set
        unvisited_nodes.remove(current_node)

        # Update the distances to neighbors of the current node
        for neighbor, weight in graph[current_node].items():
            potential_route = shortest_distances[current_node] + weight
            if potential_route < shortest_distances[neighbor]:
                shortest_distances[neighbor] = potential_route
                previous_nodes[neighbor] = current_node

    return shortest_distances, previous_nodes

# Create a sample directed graph with 6 nodes
graph = {
    'A': {'B': 2, 'C': 4, 'D': 1},
    'B': {'C': 3, 'E': 2},
    'C': {'D': 5, 'E': 1},
    'D': {'F': 3},
    'E': {'F': 7},
    'F': {}
}

# Specify the starting and ending nodes
start_node = 'A'
end_node = 'F'

# Calculate shortest distances and previous nodes using Dijkstra's algorithm
shortest_distances, previous_nodes = dijkstra(graph, start_node)

# Extract the shortest path
path = []
current = end_node
while current != start_node:
    path.insert(0, current)
    current = previous_nodes[current]
path.insert(0, start_node)

# Print the shortest path
print("Shortest Path:")
print(" -> ".join(path))

# Create a directed graph visualization
G = nx.DiGraph()

for node, edges in graph.items():
    for neighbor, weight in edges.items():
        G.add_edge(node, neighbor, weight=weight)

# Calculate the positions of nodes for visualization
pos = nx.spring_layout(G)

# Define colors for the start and end nodes
node_colors = ['green' if node == start_node else 'red' if node == end_node else 'lightblue' for node in G.nodes()]

# Create a list of edge colors where the shortest path edges are in red, and others are in light gray
edge_colors = ['red' if (u, v) in zip(path, path[1:]) else 'lightgray' for u, v in G.edges()]

# Draw the graph with edge labels and specified node and edge colors
nx.draw(G, pos, with_labels=True, node_size=500, node_color=node_colors, font_size=10, edge_color=edge_colors, arrowsize=20)
edge_labels = {(i, j): w['weight'] for i, j, w in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)

# Show the graph
plt.title("Directed Graph Visualization with Shortest Path Highlighted")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def find_cycles(graph):
    cycles = list(nx.simple_cycles(graph))
    return cycles

# Create a sample directed graph with cycles
graph_with_cycles = nx.DiGraph()
graph_with_cycles.add_edges_from([
    ('A', 'B'),
    ('B', 'C'),
    ('C', 'D'),
    ('D', 'A'),
    ('E', 'F'),
])

# Find cycles in the graph
cycles_in_graph_with_cycles = find_cycles(graph_with_cycles)

# Create a subgraph containing only the cycles
subgraph_cycles = nx.DiGraph()
for cycle in cycles_in_graph_with_cycles:
    for i in range(len(cycle)):
        subgraph_cycles.add_edge(cycle[i], cycle[(i + 1) % len(cycle)])

# Draw the subgraph containing cycles
pos = nx.spring_layout(subgraph_cycles, seed=42)
nx.draw(subgraph_cycles, pos, with_labels=True, node_size=500, font_size=10, arrowsize=20)

# Show the subgraph with cycles
plt.title("Subgraph Visualization with Cycles")
plt.show()


























```
