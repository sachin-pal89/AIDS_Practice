# Graph Data

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty graph
graph = nx.Graph()

# Add nodes
graph.add_nodes_from([1, 2, 3, 4, 5])

# Add edges
graph.add_edges_from([(1, 2), (1, 3), (2, 3), (2, 4), (3, 4), (4, 5)])

# Visualize the graph
nx.draw(graph, with_labels=True, node_color='lightblue', edge_color='gray')
plt.show()
```

```python
# Calculate degree centrality
degree_centralities = nx.degree_centrality(graph)

# Print degree centrality for each node
for node, centrality in degree_centralities.items():
    print(f"Node {node}: Degree Centrality = {centrality}")
```

Conclusion:
We can find the degree of centrality of the plotted graph.

# Spatial Data (Voice Call Quality Customer Experience India)

```python
file_path = '/content/drive/MyDrive/SEM 7/AIDS/MyCall_2022.csv'
ds = pd.read_csv(file_path, encoding='latin-1')
ds
```

```python
# Keep only the desired columns
columns_to_keep = ['latitude', 'longitude', 'rating']
ds = ds[columns_to_keep]

# Display the updated dataset
print(ds)
```

```python
import folium

df = ds

# Create a map centered around India
map = folium.Map(location=[18.519782, 73.856190], zoom_start=5)

for index, row in df.iterrows():
    if row['rating'] == 1:
        marker_color = 'blue'
    else:
        marker_color = 'red'
    folium.Marker(location=[row['latitude'], row['longitude']], icon=folium.Icon(color=marker_color)).add_to(map)

# Display the map
map
```

Conclusion:
Hence we plotted the locations in India where the Voice Call Quality Customer Experience rating is 1.
